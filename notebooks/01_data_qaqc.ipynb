```python
from src.io_utils import read_csv_robust
from src.geostats_pipeline.qaqc import run_qaqc

with open("config/project.json", "r", encoding="utf-8") as handle:
    cfg = json.load(handle)

data_path = cfg["data_csv_path"]
df = read_csv_robust(data_path)

# Ajusta estas columnas según tu dataset
collar_col = cfg.get("compositing", {}).get("alonghole_col", "BHID")
from_col = "FROM"
to_col = "TO"
azimuth_col = "AZIMUTH"
dip_col = "DIP"

metrics = run_qaqc(
    df,
    collar_col=collar_col,
    from_col=from_col,
    to_col=to_col,
    azimuth_col=azimuth_col,
    dip_col=dip_col,
    tolerance=0.0,
)

pd.DataFrame([metrics])
```
