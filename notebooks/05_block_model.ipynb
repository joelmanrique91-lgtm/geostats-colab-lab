# 05 · Block model y discretización

Este notebook muestra cómo discretizar bloques en sub‑celdas para kriging de bloque.

```python
import os, sys
import pandas as pd

PROJECT_ROOT = os.path.abspath(os.path.join(os.getcwd(), '..'))
if os.path.basename(os.getcwd()) == 'notebooks':
    os.chdir(PROJECT_ROOT)
else:
    PROJECT_ROOT = os.getcwd()
    os.chdir(PROJECT_ROOT)

if PROJECT_ROOT not in sys.path:
    sys.path.insert(0, PROJECT_ROOT)

from src.grid import grid_from_extents, make_grid_dataframe
from src.geostats_pipeline.block_model import BlockDiscretization, discretize_blocks

df = pd.read_csv('data_sample/sample_drillholes.csv')
df = df.rename(columns={'X': 'x', 'Y': 'y', 'Z': 'z'})
df.head()
```

## 1. Definir grid y discretización

```python
grid_spec = grid_from_extents(df, 'x', 'y', 'z', dx=50, dy=50, dz=10, pad=0)
grid_df = make_grid_dataframe(grid_spec)
grid_df.head()
```

```python
block_size = (grid_spec['dx'], grid_spec['dy'], grid_spec['dz'])
discretization = BlockDiscretization(nx=3, ny=3, nz=2)
subcells_df = discretize_blocks(grid_df.head(2), block_size, discretization)
subcells_df.head(10)
```

Se generaron sub‑celdas para cada bloque. El `block_id` permite asociarlas al bloque original.
